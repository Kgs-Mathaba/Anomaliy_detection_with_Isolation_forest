```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import os
from datetime import datetime
import pandas as pd
import plotly.express as px

from sklearn.ensemble import IsolationForest

mpl.rcParams['figure.figsize'] = (15, 12)
mpl.rcParams['axes.grid'] = False

import warnings
warnings.filterwarnings('ignore')
```

```python
# The dataset gives the number of taxi rides in New York city every 30 minutes
df = pd.read_csv('https://raw.githubusercontent.com/numenta/NAB/master/data/realKnownCause/nyc_taxi.csv')
```

```python
df.head(10)
```

```python
df.info()
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
df.head()
```

```python
df = df.set_index('timestamp').resample('H').mean().reset_index() # change to hourly dataset
```

```python
df
```

```python
fig = px.line(df.reset_index(), x = 'timestamp', y = 'value', title = 'NYC Taxi Demand')

fig.update_xaxes(
    rangeslider_visible = True,
)
fig.show()
```

```python
df['hour'] = df.timestamp.dt.hour
```

```python
df['weekday'] = pd.Categorical(df.timestamp.dt.strftime('%A'), categories=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
```

```python
df
```

```python
df.info()
```

## Check the Seasonality of the dataset

```python
df[['value', 'hour']].groupby('hour').mean().plot()
```

```python
df[['value','weekday']].groupby('weekday').mean().plot()
```

```python
df.groupby(['hour', 'weekday']).mean()['value'].unstack().plot()
```

```python
df.join(df.groupby(['hour', 'weekday'])['value'].mean(), on = ['hour', 'weekday'], rsuffix = '_avg')
```

```python
df_final = df.join(df.groupby(['hour', 'weekday'])['value'].mean(), on = ['hour', 'weekday'], rsuffix = '_avg')
```

```python
df_final
```

## There are holidays and event days that could be anomalous

```python
df_final.set_index('timestamp').loc['2014-11-02'] # date of NYC Marathon
```

```python
df_final.set_index('timestamp').loc['2015-01-26':'2015-01-27'] # on these days there was a blizzard
```

```python
df_final['day'] = df.timestamp.dt.weekday # numerical day value
df_final
```

```python
data = df_final[['value', 'hour', 'day']]
```

```python
model = IsolationForest(contamination=0.005, max_features=3, max_samples=0.8, n_estimators=200)
model.fit(data)
```

```python
df_final['outliers'] = pd.Series(model.predict(data)).apply(lambda x: 'yes' if (x == -1) else 'no')
```

```python
df_final.query('outliers=="yes"')
```

## Anomalous Points
- NYC Marathon - 2014-11-02
- Thanksgiving - 2014-11-27
- Christmas - 2014-12-25
- New Years - 2015-01-01
- Snow Blizzard - 2015-01-26 and 2015-01-27

```python
df_final
```

```python
fig = px.scatter(df_final, x = 'timestamp', y = 'value',color='outliers', hover_data=['weekday', 'hour', 'value_avg'],title = 'NYC Taxi Demand')

fig.update_xaxes(
    rangeslider_visible = True,
)
fig.show()
```

```python
score = model.decision_function(data)
score
```

```python
plt.hist(score, bins=50);
```

```python
df_final['scores'] = score
```
